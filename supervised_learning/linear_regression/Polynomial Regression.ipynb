```python
# TODO: Add import statements -- DONE
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Assign the data to predictor and outcome variables
# TODO: Load the data -- DONE
train_data = pd.read_csv('polynomial_regression.csv')
X = train_data['Var_X'].values.reshape((-1, 1))
y = train_data['Var_Y'].values
```

```python
# See the data
plt.scatter(X, y);
```

```python
# Create polynomial features
# TODO: Create a PolynomialFeatures object, then fit and transform the
# predictor feature -- DONE
def poly_data (X, y, degrees=2):
    poly_feat = PolynomialFeatures(degrees)
    return poly_feat.fit_transform(X, y)

def linear_regression(X, y):    
    poly_model = LinearRegression()
    poly_model.fit(X, y)
    return poly_model
```

```python
# Make and fit the polynomial regression model
# TODO: Create a LinearRegression object and fit it to the polynomial predictor
# features -- DONE
linear_data = poly_data(X, y, 1)
linear_model = linear_regression(linear_data, y)

quadratic_data = poly_data(X, y)
quadratic_model = linear_regression(quadratic_data, y)

cubic_data = poly_data(X, y, 3)
cubic_model = linear_regression(cubic_data, y)

quadrilateral_data = poly_data(X, y, 4)
quadrilateral_model = linear_regression(quadrilateral_data, y)
```

```python
# Draw the linear model
plt.scatter(X, y) # Data
plt.plot(X, linear_model.predict(linear_data), 'r+');
```

```python
# Draw the quadratic model
plt.scatter(X, y) # Data
plt.plot(X, quadratic_model.predict(quadratic_data), 'r^');
```

```python
# Draw the cubic model
plt.scatter(X, y) # Data
plt.scatter(X, cubic_model.predict(cubic_data), color='black');
```

```python
# Draw the 4D model
plt.scatter(X, y) # Data
plt.scatter(X, quadrilateral_model.predict(quadrilateral_data), color='green');
```
